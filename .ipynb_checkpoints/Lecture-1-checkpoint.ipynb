### Categorial Variables

**To summarize**, we report the counts of each possible category.

**To visualize**, we make a bar plot.

````python
titanic_df["pclass"].value_counts()
````

````python
titanic_df["pclass"].value_counts().sort_index().plot.bar()
# If we don't sort index it will plot sorted with values (desc)
````

### Handling Two or More Categorial Variables

````python
# One variable

titanic_df["pclass"]
````

The result is a one-dimensional pandas object called a `Series`

To select multiple columns we should pass a **list** of colunm names.

````python
titanic_df[["pclass", "survived"]]
````

This time the result is two-dimensional, another smaller `DataFrame`

To Summarize multiple categorial variables, we report the counts of **every possible combination** of catwgories

````python
titanic_df[["pclass", "survived"]].value_counts()
````

Note that the result is a `series`, with a multi-level index, one for each variable!

To make data more readable `.unstack()` can be used. This representation is called a **two-way table** or a **crosstab** (short for "cross-tabulation").

````python
titanic_crosstab = titanic_df[["pclass", "survived"]].value_counts().unstack()

titanic_crosstab
````

From a crosstab, we can make a bar plot to visualize the data. This type of plot is called a `Grouped bar plot`


````python
titanic_crosstab.plot.bar()
````

We can recover the counts for each individual variable from a crosstab

* We can sum over the columns (across each row) to obtain the counts for `pclass`
* Or we can sum over the rows (down each column) to obtain counts for `survived`

````python
pclass_count = titanic_crosstab.sum(axis="columns")
survived_count = titanic_crosstab.sum(axis="rows")

pclass_count, survived_count
````

### Proportions

Instead of counts, it can be useful to report *proportions*, where we normalize by the total.

`proportion = count / total`

**Example:** The proportions of the three passanger classes

````python
pclass_distribution = titanic_df["pclass"].value_counts() / len(titanic_df)

pclass_distribution
````

Together, the proportions of a categorical variable are called the **distribution** of the variable.

! Notice that the values in a distribution add up to 1.0

````python
pclass_distribution.sum()
````

### Probablities

- What does it mean to say, “The proportion of passengers in 3rd class is 0.541635?”
- One interpretation is as a `probability`.
- “If we were to pick a passenger on the Titanic at random, the probability that they are in 3rd class is 0.541635.”
- We notate this as `P(3rd class) = 0.541635`
- If we want to be explicit about the variable and the category, `P(pclass = 3) = 0.541635`

### Vectorization

! In Pandas, operations are **vectorized**. A **series** behaves like a vector. Vectors in pandas work like vectors in math!

So we can divide a series by a number.

Ex: ``` df["pclass"].value_counts() / len(df) ```

### Joint Distributions

If we calculate the distribution of multiple variables, this is called a `Joint Distribution`

````python
joint = titanic_df[["pclass", "survived"]].value_counts().unstack() / len(titanic_df)

joint
````

Notice that the values in the joint distribution also sum up to 1.0

````python
joint.sum().sum()
````

When we visualize a joint distribution, the y-axis scale changes, but the shape keeps same.

To appreciate the power of proportions, we need to look at **conditional distributions**.

````python
joint.sort_index().plot.bar()
````

### Conditional Probablities

- What does it mean to say, “The conditional proportion of survival given 3rd class is 0.255289”?
- One interpretation is as aconditional probability.
- “If we were to pick a 3rd class passenger on the Titanic at random, the probability that they survived is **0.255289**.”
- We notate this as `P(survived|3rd class) = 0.255289`
