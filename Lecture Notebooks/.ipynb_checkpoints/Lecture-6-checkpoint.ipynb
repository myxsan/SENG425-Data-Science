# Distances Between Observations

In this notebook we will use the dataset Houses in Ames, IA

- **Gr Liv Area:** Ground living area
- **Bedroom AbvGr:** Bedroom above ground
- **Mo Sold:** Month sold
- **WD** Waranty deed

```python
import pandas as pd

df = pd.read_table("https://datasci112.stanford.edu/data/housing.tsv")
df
```

We've already seen how to visualize and summarize the relationship between quantiative variables.

```python
df.plot.scatter(x="Gr Liv Area", y="Bedroom AbvGr", alpha=.4)
df[["Gr Liv Area", "Bedroom AbvGr"]].corr()
```

## Quantifying Similarity
How do we measure how "smilar" two observations are?

One way is to calculate the distance between two points.

**Distance = 0:** Houses have almost identical features.
**Moderate distance:** Houses are similar but with some differences.
**Very Large distance:** Houses differ greatly in size or room counts.

## Variable Scaling

```python
df.plot.scatter(x="Gr Liv Area", y="Bedroom AbvGr", alpha=.4) # This plot is fake
```

```python
# This is how it really looks
df.plot.scatter(x="Gr Liv Area", y="Bedroom AbvGr", alpha=.4, ylim=(0, 5000)) 
```

The variation in square footage completely swamps the variation in the number of bedrooms.

We should bring all the variables to the same scale before calculating distance.

One way of doing this is called **Standardization**

**Standardization:** A type of scaling, similar to normalization using mean and standart devation instead of compressing values into a fixed range.

Let's obtain a new DataFrame where each column has a **mean of 0** and a **standart deviation of 1**.

```python
features = ["Gr Liv Area", "Bedroom AbvGr", "Full Bath", "Half Bath"]
df_standardized = ((df[features] - df[features].mean(axis="rows")) / 
                   df[features].std(axis="rows"))

df_standardized
```

In the standardized version
- Each feature has mean = 0
- Each feature has standart deviation = 1
- All features are on the same scale.
- Prevents features with larger ranges from dominating analyses.

In this way every variable is treated equally.

```python
df_standardized.plot.scatter(x="Gr Liv Area", y="Bedroom AbvGr", alpha=0.4)
```

Now we can find the most similar house by calculating distances on the **standardized** data.

Let's start with computing the difference between each row and row 1707.

1. Square the differences and sum them across columns.
2. Take the square roo to get Euclidean distances.
3. Sort distances from closest to farthest.

```python
import numpy as np

diffs = df_standardized - df_standardized.loc[1707]
dists = np.sqrt((diffs ** 2).sum(axis="columns"))

dists.sort_values()
```

The most similar house to **#1707** is **#160**, which matches our intuition when we looked at the table.

## Calculating Distances(Scaling) in Scikit-Learn

The machine learning library Scikit-Learn can be used to scale variables and calculate distances.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# This calculates the meand and std of each variable.
scaler.fit(df[features])

# This actually does the standardization.
array_standardized = scaler.transform(df[features])
array_standardized # Scaled numerical array
```

The result is same as before, but it's in an array, not a `DataFrame`

Now we can calculate the distance between **#1707** and all houses, using Scikit-Learn

```python
from sklearn.metrics import pairwise_distances

dists = pairwise_distances(array_standardized[[1707]], array_standardized) # Calculate distances
dists[0].sort() # Sort distances

dists
```

Now we can see values but we can't see which value is which house's. In this case we can use `argsort()`.

- `argsort()` returns the indices that would sort an array or Series.
- The smallest values' indices come first, largest last.
- Useful for ranking or finding neares/farthest items without changing the original array.

DO NOT USE `argsort()` after `sort()`. So recalculated in the next cell.

```python
dists = pairwise_distances(array_standardized[[1707]], array_standardized) # Calculate distances
dists[0].argsort() # Sort distances
```

## The Distance Matrix

We can calculate the distances every pair of observations and store the result in a matrix.

```python
dist_matrix = pairwise_distances(array_standardized)

dist_matrix
```

Note that row #1707 of this matrix contains the *dists* we calculated previously.
