# Machine Learning

### What is ML

**Classic AI:** Attempts to codify the rules that a human would use to make decisions.

*Example:* If you are trying to build a system that finds all the proper nouns in a text document, yoou might **hard-code** the following rules.

- If a word is a proper noun, then the first letter of the word is capitalized
- The first letter of a sentence is always capitalized.
- ...

**The system can deduce new rules from existing rules**
- It is impossible to tell whether the first word of a sentence is a proper noun just from capitalization.


**Essence of ML:** Rather than trying to come up with the rules ourselves, we can learn from data.

**Learning:** The act of coming up with a rule for making decisions based on a set of inputs.

Inputs (x) -------▶ |f| -------▶ Decision (y)


Goal of Machine Learning: Come up with a rule f from training data (xᵢ, yᵢ)

The decision called typically the **target** or the **label**

```python
import pandas as pd

df = pd.read_csv("https://dlsun.github.io/pods/data/bordeaux.csv", index_col="year")
df.head()
```

**Summer:** Average summer temperature(°C). Affects grape ripening and wine quality.

**Har(Harvest Rainfall):** Rainfall during harvest (mm). Too much rain can reduce quality.

**Sep:** Average September temperature. Important for final grape ripening.

**Wing:** Winter rainfall. Influences soil water and vine growth.

**Age:** Age of the wine, Older wines are usually more valuable.

summer, har, sep, win, age (x) -------▶ |f| -------▶ price (y)


```python
import plotly.express as px

fig1 = px.scatter(df[~df["price"].isnull()],
           x="win", y="summer", color="price")

fig1
```

plotly.express = fast and easy way to create
plotsplotly.graph_objects = more detailed, low-level, and high customizable plotting.

```python
import plotly.graph_objects as go

fig2 = px.scatter(df[df["price"].isnull()],
                  x="win", y="summer", symbol_sequence=["circle-open"], hover_name=df[df["price"].isnull()].index)

go.Figure(data = fig1.data + fig2.data, layout=fig1.layout)
```

**Insgight:** The closest wines are low quality, so the 1986 vintage is probably low quality as well.

**This is the intuition behind k-nearest neighbors**

### Types of Mechine Learning Problems

Machine learning problems are grouped into two types, based on the type of $y$

- **Regression:** The labe $y$ is quantitative.
- **Classification:** The label $y$ is categorical.

Note that the input $x$ may be categorical, quantitative, textual, ..., or any combination of these.
