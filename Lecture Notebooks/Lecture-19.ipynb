# Combinin Data Sets

Information may be spread across multiple data sets. 

For example, suppose we want to know which manufacturer's planes made the most flights in November 2013. One data set contains information about flights in Nov. 2013 while another contains information about planes.

```python
import pandas as pd
data_dir = "https://datasci112.stanford.edu/data/nycflights13/"
```

```python
df_flights = pd.read_csv(f"{data_dir}/flights11.csv")
df_flights
```

```python
df_planes = pd.read_csv(f"{data_dir}/planes.csv")
df_planes
```

### Keys

**Primary Key:** A column (or a set of columns) that uniquely identifies observations in adata frame.

**Foreign Key:** A column (or a set of columns) that points to the primary key of another data frame

## Joining on a Key

- The Pandas funciton `.merge()` can be used to join two `DataFrame`s on a key.

- Joining two data frames results in a *wider* data frame, with more columns.

```python
df_joined = df_flights.merge(df_planes, on="tailnum")
df_joined
```

### Overlapping Column Names

- In our example both data frames contain a column named "year", but we did not joing on this as a key.
- By default, Pandas adds the suffixes _x and _y to overlapping column names, but this can be customized

```python
df_joined = df_flights.merge(df_planes, on="tailnum",
                             suffixes=("_flight", "_plane"))
df_joined
```

## Joining on Multiple Keys

- In our example, we can also join the flight data to weather data.

```python
df_weather = pd.read_csv(f"{data_dir}/weather.csv")
df_weather
```

#### A key with Multiple Columns

In weather we don't have unique primary key in one column. We need to join to the weather data on year, month, day, and hour.

`df_flights[df_flights["origin"] == "JFK"]` -> Flights only in JFK airport.

`df_weather[df_weather["airport"] == "JFK"]` -> Weather data for only JFK airport.

`on=("year", "month", "day", "hour")` -> Flights and weather are merged at matching time.

`df_jfk` -> Merged data set.

```python
df_jfk = df_flights[df_flights["origin"] == "JFK"].merge(
    df_weather[df_weather["airport"] == "JFK"],
    on=("year", "month", "day", "hour")
)

df_jfk
```

```python
df_jfk.plot.scatter(x="visib", y="dep_delay", alpha=0.2)
```

Let's see how average departure delay changes with visibility.

- `groupby("visib")` -> Groyps the data by visibility.
- `["dep_delay"].mean()` -> Computes the mean departure delay for each visibility level.
- `plot.line()` -> Displays the result as a line plot.

```python
df_jfk.groupby("visib")["dep_delay"].mean().plot.line()
```

### Joining on Keys with Different Names

- Sometimes, the join keys have different names in the two data sets. This happens if the data sets come from different sources.
- In our example, if we want to join the entire flights data to the weather data, we would need to include the airport in the key but the airport is called "origin" in df_flights and "airport" in df_weather.
- The `.merge()` function provides `left_on=` and `right_on` arguments for specifying different column names in the **left** and **right** data frames.

```python
df_flights_weather = df_flights.merge(
    df_weather,
    left_on=("origin", "year", "month", "day", "hour"),
    right_on=("airport", "year", "month", "day", "hour")
)

df_flights_weather
```

Notice that df_flights_weather contains columns called origin and airport.

Now that the data is joined, we can visualize the average departure delay as a function of the visibility at the three airports.

```python
#To compare how the average departure delay changes with visibility for each airport.
(df_flights_weather
 .groupby(["airport", "visib"])["dep_delay"].mean()
 .unstack("airport")
 .plot.line())
```
